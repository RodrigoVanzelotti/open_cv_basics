```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show_img(titulo, image):
    cv2.imshow(titulo, image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

### Histogramas
Um histograma representa a distribuio de intensidades de pixel (seja colorido ou em tons de cinza) em uma imagem. 
Ele pode ser visualizado como um grfico (ou grfico) que fornece uma intuio de alto nvel da distribuio de intensidade (valor de pixel).
Vamos assumir um espao de cores RGB neste exemplo, ento esses valores de pixel estaro no intervalo de 0 a 255.







```python
# processo padro de ler a imagem e criar um canva vazio
img = cv2.imread('../assets/fotos/cats.jpg')
# img = cv2.imread('../assets/fotos/cat.jpg')
show_img('gatitos da asimov', img)

blank = np.zeros(img.shape[:2], dtype='uint8')
```

```python
# primeiro, vamos calcular o histograma da imagem em cinza

# converter a imagem para P&B
cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
show_img('gatitos da asimov', cinza)
```

```python
# Histograma de P&B
# cv2.calcHist(imagem, canais_de_cor(RGB, BGR), mascara, tamanho_do_hist, ranges)
cinza_hist = cv2.calcHist([cinza], [0], None, [256], [0, 256])

# plotando a imagem desse nosso histograma bsico das escalas de cinza
```

Ao plotar essa figura, podemos observar que temos uma grande concentrao de
valores prximos de zero, afinal estamos observando uma imagem P&B, onde o
0 representa valores completamente escuros e 255 valores brancos

```python
plt.figure()
plt.title('Histograma de escalas de cinza, sem mscara')
plt.xlabel('Intervalos/Setores')
plt.ylabel('Nmero de Pixels')
plt.plot(cinza_hist)
plt.xlim([0,256])
plt.show()
```

```python
# vamos rever a imagem na prxima clula para ter uma melhor compreenso
show_img('gatitos da asimov', cinza)
```

Neste caso, como voltamos a imagem original (no a preto e branco), podemos analisar canais diferentes alm do preto ao branco, os canais RGB (lidos pelo cv2 como BGR).
Para isso, iteraremos pelos canais, traando novas linhas em nosso grfico.

```python
plt.figure()
plt.title('Histograma de cores, sem mscara')
plt.xlabel('Intervalos/Setores')
plt.ylabel('Nmero de Pixels')
colors = ['b', 'g', 'r']
for channel, cor in enumerate(colors):
    hist = cv2.calcHist([img], [channel], None, [256], [0,256])
    plt.plot(hist, color=cor)
    plt.xlim([0,256])

plt.show()
```

Se quisermos, podemos analisar trechos especficos de cada imagem, aplicando apenas uma mscara na imagem e traando o histograma (para plotar  o mesmo cdigo, porm com a adio da mscara)

```python
mask = cv2.circle(blank, (img.shape[1]//2,img.shape[0]//2), 100, 255, -1)
show_img('Mascara', mask)

masked = cv2.bitwise_and(img, img, mask=mask)
show_img('Mascara na imagem', masked)
```

```python
plt.figure()
plt.title('Histograma de cores, sem mscara')
plt.xlabel('Intervalos/Setores')
plt.ylabel('Nmero de Pixels')
colors = ['b', 'g', 'r']
for channel, cor in enumerate(colors):
    hist = cv2.calcHist([img], [channel], mask, [256], [0,256])
    plt.plot(hist, color=cor)
    plt.xlim([0,256])

plt.show()
```
